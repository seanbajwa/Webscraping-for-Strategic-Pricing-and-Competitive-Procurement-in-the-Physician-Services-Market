```python
def getCodes(file):
    codes = set()
    with open(file) as f:
        for line in f:
            if len(line) > 0:
                line = line.strip()
                codes.add(line)
    return codes
```

```python
code_file = '/Users/seanbajwa/Desktop/data_revival/hcpc_codes.txt'
```

```python
codes = list(getCodes(code_file))
```

```python
len(codes)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
new_root = root+ 'data/'
```

```python
years = [year for year in os.listdir(new_root) if(year[0].isdigit())]
```

```python
year_len = len(years)
```

```python
code_dict = {}
for code in codes:
    year_list = [None]*(year_len)
    for i in range(len(years)):
        year = years[i]
        curr_df = pd.read_csv(new_root + year)
        code_list = list(set(curr_df['HCPCS CODE'].tolist()))
        if(int(code) in code_list):
            year_list[i] = True
        else:
            year_list[i] = False
    code_dict[code] = year_list    
```

```python
str_years = [str([year[:-4]]) for year in years]
```

```python
data = []
for key in code_dict:
    key_list = code_dict[key]
    row = [key] + key_list
    data.append(row)
```

```python
str_years = ['HCPCS CODE'] + str_years
bool_df = pd.Dataframe(data, columns=str_years)
bool_df.head()
```
